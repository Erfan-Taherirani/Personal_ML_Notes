# **Encoding Strategies for Categorical Variables**:

- Label Encoding
- Ordinal Encoding
- One-hot Encoding
- Target Encoding
- Frequency Encoding
- Binary Encoding
- Embedding Encoding

## **Label Encoding**:

This encoding strategy assigns an integer to every unique category in the categorical variable.
> Note: Mostly used for encoding the target variable in the classification task.

## **Ordinal Encoding**:

This encoding strategy used for encoding ranked qualitative data and when there is a meaninful ranking between categories in a variable.

example: bachelor, master, PHD

## **One-hot Encoding**:

This method can encode categorical variables without introducing any order between them.
> Note: Resulting in high dimensional feature space is the problem of this encoding strategy (Curse of dimensionality).

```python
diamonds = sns.load_dataset("diamonds")
print(f"Shape of the dataset: {diamonds.shape}")
diamonds.head(10)
```

```python
print("Data types in the dataset:")
diamonds.dtypes
```

```python
diamonds['cut'].unique()
```

```python
import pandas as pd
import numpy as np

enc = OneHotEncoder(sparse_output=False)
cut_encoded = enc.fit_transform(diamonds['cut'].to_numpy().reshape(-1, 1))
cut_encoded = pd.DataFrame(
    data=cut_encoded,
    columns=enc.get_feature_names_out(input_features=["cut"]),
    dtype=np.int8
)

cols = diamonds.columns.to_list()
cols.pop(1)
diamonds = pd.concat([diamonds[cols], cut_encoded], axis=1)
```

```python
import pandas as pd

np.random.seed(42)
df = pd.DataFrame({
    'City': np.random.choice(['New York', 'London', 'Paris', 'Tokyo'], 1000),
    'Sales': np.random.normal(loc=100, scale=20, size=1000)
})

X = df['City'].to_numpy().reshape(-1, 1)
y = df['Sales'].to_numpy().reshape(-1, 1)

# encoding
one_hot_enc = OneHotEncoder(sparse_output=False)
city_onehot_encoded = one_hot_enc.fit_transform(X)
df[["City_London", "City_New York", "City_Paris", "City_Tokyo"]] = city_onehot_encoded
df
```

## **Target Encoding**:

This encoding method uses the target variable to show how each category influences the target variable
> Note: Useful for categorical variables with high cardinality cause doesn't make more than one feature. It prevents the curse of dimensionality.

```python
X = df['City'].to_numpy().reshape(-1, 1)
y = df['Sales'].to_numpy().reshape(-1, 1)

target_encoder = TargetEncoder()
city_target_encoded = target_encoder.fit_transform(X, y)
df["City_target_encoded"] = city_target_encoded
df
```

## **Frequency Encoding**:

This method use the frequency of categories in cout or percent mode.
> Note: This encoding method is also useful for categorical variables with high cardinality.

```python
def frequency_encoding(s: pd.Series) -> pd.Series:
    mapping = s.value_counts() / len(s)
    return s.map(mapping)
```

```python
df['City_frequency'] = frequency_encoding(df['City'])
df
```

## **Binary Encoding**:

This encoding method uses binary representation of the categories and can represent the categorical variable and input it to the computer but in may introduce some order between categories and introduces bias to the model.
